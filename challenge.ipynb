1. Definição das Estruturas e Funções
Nesta seção, definimos todas as classes (Fila, Pilha) e funções (buscas, ordenação) que serão utilizadas na simulação.

```python
# Célula 1: Definições de Dados, Classes e Funções
# =======================================================

import datetime
from collections import deque

# --- Conteúdo que estava em 'insumo.py' ---
dados_consumo = [
    {'id': 1, 'nome': 'Tubo de Ensaio', 'quantidade': 150, 'data_consumo': datetime.date(2025, 9, 10), 'validade': datetime.date(2026, 8, 1)},
    {'id': 2, 'nome': 'Reagente A', 'quantidade': 30, 'data_consumo': datetime.date(2025, 9, 10), 'validade': datetime.date(2025, 12, 15)},
    {'id': 3, 'nome': 'Luvas Descartáveis', 'quantidade': 300, 'data_consumo': datetime.date(2025, 9, 11), 'validade': datetime.date(2027, 1, 1)},
    {'id': 4, 'nome': 'Agulha Hipodérmica', 'quantidade': 200, 'data_consumo': datetime.date(2025, 9, 11), 'validade': datetime.date(2026, 11, 20)},
    {'id': 5, 'nome': 'Reagente B', 'quantidade': 25, 'data_consumo': datetime.date(2025, 9, 12), 'validade': datetime.date(2025, 11, 30)},
    {'id': 6, 'nome': 'Tubo de Ensaio', 'quantidade': 120, 'data_consumo': datetime.date(2025, 9, 12), 'validade': datetime.date(2026, 8, 1)},
    {'id': 7, 'nome': 'Reagente A', 'quantidade': 35, 'data_consumo': datetime.date(2025, 9, 13), 'validade': datetime.date(2025, 12, 15)},
    {'id': 8, 'nome': 'Seringa 5ml', 'quantidade': 250, 'data_consumo': datetime.date(2025, 9, 13), 'validade': datetime.date(2027, 5, 10)},
    {'id': 9, 'nome': 'Reagente C', 'quantidade': 15, 'data_consumo': datetime.date(2025, 9, 14), 'validade': datetime.date(2026, 2, 28)},
    {'id': 10, 'nome': 'Agulha Hipodérmica', 'quantidade': 180, 'data_consumo': datetime.date(2025, 9, 14), 'validade': datetime.date(2026, 11, 20)},
]

# --- Conteúdo que estava em 'fila_pilha.py' ---
class FilaDeConsumo:
    def __init__(self):
        self.fila = deque()

    def registrar_consumo(self, insumo):
        print(f"Registrando na fila: {insumo['nome']} (Qtd: {insumo['quantidade']})")
        self.fila.append(insumo)

    def processar_proximo(self):
        if not self.fila:
            return "Fila de consumo vazia."
        return self.fila.popleft()

class PilhaDeConsultas:
    def __init__(self):
        self.pilha = []

    def adicionar_para_consulta(self, insumo):
        self.pilha.append(insumo)

    def consultar_ultimo_consumo(self):
        if not self.pilha:
            return "Pilha de consultas vazia."
        return self.pilha.pop()

# --- Conteúdo que estava em 'buscas.py' ---
def busca_sequencial(lista, nome_insumo):
    encontrados = []
    print(f"\n... Executando Busca Sequencial por '{nome_insumo}' ...")
    for registro in lista:
        if registro['nome'].lower() == nome_insumo.lower():
            encontrados.append(registro)
    return encontrados

def busca_binaria(lista_ordenada, nome_insumo):
    print(f"\n... Executando Busca Binária por '{nome_insumo}' ...")
    inicio, fim = 0, len(lista_ordenada) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        nome_meio = lista_ordenada[meio]['nome'].lower()
        if nome_meio == nome_insumo.lower():
            return lista_ordenada[meio]
        elif nome_insumo.lower() < nome_meio:
            fim = meio - 1
        else:
            inicio = meio + 1
    return None

# --- Conteúdo que estava em 'ordenacao.py' ---
def merge_sort(lista, chave='quantidade'):
    if len(lista) <= 1:
        return lista
    meio = len(lista) // 2
    esquerda = merge_sort(lista[:meio], chave)
    direita = merge_sort(lista[meio:], chave)
    # Função _merge auxiliar
    def _merge(esquerda, direita, chave):
        resultado = []
        i = j = 0
        while i < len(esquerda) and j < len(direita):
            if esquerda[i][chave] <= direita[j][chave]:
                resultado.append(esquerda[i])
                i += 1
            else:
                resultado.append(direita[j])
                j += 1
        resultado.extend(esquerda[i:])
        resultado.extend(direita[j:])
        return resultado
    return _merge(esquerda, direita, chave)

def quick_sort(lista, chave='validade'):
    if len(lista) <= 1:
        return lista
    else:
        pivo = lista[len(lista) // 2]
        menores = [x for x in lista if x[chave] < pivo[chave]]
        iguais = [x for x in lista if x[chave] == pivo[chave]]
        maiores = [x for x in lista if x[chave] > pivo[chave]]
        return quick_sort(menores, chave) + iguais + quick_sort(maiores, chave)

# --- Conteúdo que estava em 'analise_hash.py' ---
def agregar_consumo_por_insumo(lista_consumo):
    consumo_agregado = {}
    for registro in lista_consumo:
        nome = registro['nome']
        quantidade = registro['quantidade']
        if nome in consumo_agregado:
            consumo_agregado[nome] += quantidade
        else:
            consumo_agregado[nome] = quantidade
    return consumo_agregado

# --- Função auxiliar para impressão ---
def imprimir_relatorio(titulo, dados):
    print("-" * 60)
    print(f"| Relatório: {titulo}")
    print("-" * 60)
    if not dados:
        print("| Nenhum dado para exibir.")
    elif isinstance(dados, dict):
        for chave, valor in dados.items():
            print(f"| - {chave:<25} | Total Consumido: {valor}")
    elif isinstance(dados, list):
        for item in dados:
            print(f"| - {item['nome']:<25} | Qtd: {item['quantidade']:<5} | Validade: {item['validade']}")
    else:
        print(f"| - {dados['nome']:<25} | Qtd: {dados['quantidade']:<5} | Validade: {dados['validade']}")
    print("-" * 60 + "\n")

#
# FIM DA CÉLULA 1
#
```

2. Simulação e Geração de Relatórios
Agora, executamos o script principal que utiliza as funções definidas acima para simular o registro de consumo e gerar os relatórios de análise.

```python
# Célula 2: Execução Principal (o que estava em 'main.py')
# ==========================================================
# Agora que tudo foi definido na célula anterior, podemos usar as funções e classes diretamente.

print(">>> Iniciando Simulação de Gerenciamento de Insumos <<<\n")

# 1. Registro de consumo diário usando Fila
fila_registro = FilaDeConsumo()
for consumo in dados_consumo:
    fila_registro.registrar_consumo(consumo)
print("\n")

# 2. Busca por um insumo específico (Sequencial)
insumo_procurado = 'Agulha Hipodérmica'
encontrados_seq = busca_sequencial(dados_consumo, insumo_procurado)
imprimir_relatorio(f"Busca Sequencial por '{insumo_procurado}'", encontrados_seq)

# 3. Exemplo de Busca Binária (requer lista ordenada por nome)
dados_ordenados_nome = sorted(dados_consumo, key=lambda x: x['nome'])
encontrado_bin = busca_binaria(dados_ordenados_nome, 'reagente b')
imprimir_relatorio(f"Busca Binária por 'Reagente B'", [encontrado_bin] if encontrado_bin else [])

# 4. Relatório de Insumos Mais Consumidos (ordenado com Merge Sort)
ordenados_por_qtd = merge_sort(dados_consumo, chave='quantidade')
ordenados_por_qtd.reverse() # Para mostrar do maior para o menor
imprimir_relatorio("Insumos por Quantidade Consumida (Maior para Menor)", ordenados_por_qtd)

# 5. Relatório de Controle de Validade (ordenado com Quick Sort)
ordenados_por_validade = quick_sort(dados_consumo, chave='validade')
imprimir_relatorio("Insumos por Data de Validade (Estratégia FEFO)", ordenados_por_validade)

# 6. Relatório de Consumo Total Agregado (usando Tabela Hash)
consumo_total = agregar_consumo_por_insumo(dados_consumo)
imprimir_relatorio("Consumo Total Agregado por Insumo", consumo_total)

print(">>> Simulação Finalizada <<<")
```
